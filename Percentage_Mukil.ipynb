```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##**Percentage for 2 columns**

```
df = pd.read_excel('/content/study7_trail_visual.xlsx')
```

**You can not calculate percentage changes if any of the column value is zero**

Ex- 0 to 0.17 : No percentage


0.9 to 0 no percentage


**Excluded all such values**

You can suggest something if you disagree. :)

```
df = df[df['FO-71'] > 0]

df = df[df['inulin_average'] > 0]

df = df.reset_index(drop = True)
```

```
df['percentage'] = pd.Series(df.apply(lambda row: (row.iloc[2]-row.iloc[1])/row.iloc[2]*100, axis=1))
```

## **Read and Process Survey 6 and 7**

```
df6 = pd.read_csv('/content/study6_RA.csv')
df6 = df6.T
df6.columns = df6.iloc[0]
df6 = df6.iloc[1:]

df6['Avg_Control'] = df6[df6.columns[pd.Series(df6.columns).str.startswith('Control_')]].mean(axis = 1)
df6 = df6.drop( columns = df6.columns[pd.Series(df6.columns).str.startswith('Control')])

df7 = pd.read_csv('/content/study7_RA.csv')
df7 = df7.T
df7.columns = df7.iloc[0]
df7 = df7.iloc[1:]
df7['Avg_inulin'] = df7[df7.columns[pd.Series(df7.columns).str.startswith('inulin')]].mean(axis = 1)
df7 = df7.drop( columns = df7.columns[pd.Series(df7.columns).str.startswith('inulin')])


```

## **Merging**

```
df6['Taxonomy'] = df6.index
df7['Taxonomy'] = df7.index
df = df6.merge(df7, how = 'outer', on = 'Taxonomy')
df.index = df['Taxonomy']
df.drop(columns = ['Taxonomy'], inplace= True)
df = df.astype('float')
```

## **Create Heatmap**

### **Average Control**

```
df = df.sub(df['Avg_Control'], axis=0)
```

**Here I have taken 20, which you can change to whatever you need**

```
df = df.drop(columns = ['Avg_Control', 'Avg_inulin'])
df = df[(df > 20).any(1)]
```

```
#df = df.dropna()
```

```
fig, ax = plt.subplots(figsize=(30,6))
sns_plot = sns.heatmap(df, annot=True, linewidths=.5, ax=ax);
ax.set_title('Average Control');
#plt.savefig('foo.png')
```

### **Average Inulin**

```
df = df.sub(df['Avg_inulin'], axis=0)
```

**Here I have taken 20, which you can change to whatever you need**

```
df = df.drop(columns = ['Avg_Control', 'Avg_inulin'])
df = df[(df > 20).any(1)]
```

```
fig, ax = plt.subplots(figsize=(30,6))
sns_plot = sns.heatmap(df, annot=True, linewidths=.5, ax=ax);
ax.set_title('Average Inulin');
#plt.savefig('foo.png')
```
